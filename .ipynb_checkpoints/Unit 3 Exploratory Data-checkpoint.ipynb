### Step 5 - Create Plots for Each Variable in Dataset

```python
# import seaborn
import seaborn as sns

# set standard plot type as seaborn plot
sns.set()

# import matplotlib
import matplotlib.pyplot as plt

# write command for plots to appear in jupyter cells automatically
%matplotlib inline
```

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

### Plot for 'age' column

```python
# create plot for age column
sns.histplot(data=df, x='age', bins=5)
plt.xlabel("Age")
```

### Plot for 'industry' column

```python
# create plot for industry column
data = df['industry'].value_counts()
labels= ['business', 'finance', 'technology', 'healthcare']
plt.pie(data, labels=labels, colors= sns.color_palette('viridis'), autopct = '%.0f%%')
plt.title("Industry")
```

##### Insights:

it looks like the data used the same amount of surveys from each industry

### Plot for 'salary' column

```python
# create plot for salary column
sns.boxplot(data=df, y='salary', palette = "OrRd").set(title='Manager Salaries')
```

50% of all those surveyed make between 85,000 and 120,000 per year

### Plot for 'industry_experience' column

```python
# create plot for industry experience column
sns.histplot(data=df, x='industry_experience', kde=True)
```

##### Insights:

the graph shows that the most work experience of all those surveyed is about 20 years

### Plot for 'education' column

```python
# create plot for education levels
data = df['education'].value_counts()
labels=['high school', 'some college', 'bachelors degree',
       'masters degree']
plt.pie(data, labels=labels, colors= sns.color_palette('pastel'), autopct = '%.0f%%')
plt.title('Education')
```

### Step 2 - Review Hypothesis Testing & P-Value

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

#### Fill in blanks with brief description of the following terms:

- **Hypothesis**
> a best guess of what the data is going to tell you

- **Null Hypothesis**
> there is no significant difference or change.  The only change is due to chance.

- **P-Value**
> the probability of exceeding the original mean difference. threshold typically 5%

### Step 3 - Use Pearsonr & Spearmanr To Answer Questions About Correlation and P-Values

<h3 style='color:red;'> First, load 'salary_numeric.csv' - where education has been re-expressed as numeric variables.</h3>

```python
# load salary_numeric.csv into dataframe where education levels have been re-expressed as numeric
df_num=pd.read_csv('salary_numeric.csv')
```

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

```python
# import pearson and spearmanr methods from scipy.stats library
from scipy.stats import pearsonr, spearmanr
```

### What is the correlation between salary and education?

#### Your Hypothesis = salary increases with education

```python
# calculate correlation and pvalue of salary and education with spearmanr method
spearmanr(df_num['salary'], df_num['education'])
```

There is a positive correlation between salary and education.

#### Are these results statistically significant? How do you know?

Yes, because the p-value is 0

### What is the correlation between salary and age?

#### Your Hypothesis salary increases with age

```python
# calculate correlation and pvalue of salary and age with pearsonr method
pearsonr(df_num['salary'], df_num['age'])
```

There is a positive correlation between salary and age

#### Are these results statistically significant? How do you know? 

Yes, because the p-value is 0

### What is the correlation between salary and industry experience? 

#### Your Hypothesis = salary increases with industry experience

```python
# calculate correlation and pvalue of salary and industry experience with pearsonr method
pearsonr(df_num['salary'], df_num['industry_experience'])
```

There is a positive correlation between salary and industry experience

#### Are these results statistically significant? How do you know?

Yes, because the p-value is 0

#### Crosstab of average salary grouped by education and industry

```python
# use pd.crosstab() to calculate mean salary in each industry grouped by education level
pd.crosstab(index=df_num['industry'], columns=df_num['education'], values=df_num['salary'], aggfunc=np.mean).round(2)
```

#### Heatmap of crosstab of average salary grouped by education and industry

```python
# save above result to new variable
crosstab=pd.crosstab(index=df_num['industry'], columns=df_num['education'], values=df_num['salary'], aggfunc=np.mean).round(0)
```

```python
# create heatmap with annotated mean salary in each industry and education level
plt.figure(figsize=(10,5))
sns.heatmap(crosstab, annot=True, cmap="Blues", fmt='g', xticklabels=['High School', 'Some College', "Bachelor's Degree", "Master's Degree"])
```

##### Insights:

the higher the education the higher the salary with technology being the best
